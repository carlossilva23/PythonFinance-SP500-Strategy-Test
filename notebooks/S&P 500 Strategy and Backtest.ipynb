Today my goals are to pick my short and long moving averages and create logic to automatically calculate both short and long MA's for each stock in the portfolio. After each MA's are calculated I will then create logic to create buy and sell signals based on crossovers and intersections. 

For my short moving average, I'm going to pick a 25 day MA (a little less than a month), and for my long MA I will be choosing 100 days (a little over 3 months), because I feel like these are ample enough times to create the best data. 

In order to calculate these, I'm thinking of creating a function that takes the stock dataframe as a parameter and adds a column for each stock with these values calculated. My plan is to look through the pandas docs in order to figure out how to create a new column and have it automaticaly fill each column out. 

With pandas, we can easily calculate a moving average within a certain time period with the function .rolling(). We can combine this with the mean (which literally is just an average), and then append each result to a new column. For example, this is how it works:

```python
# We first get the ticker data for a stock 
ticker = yf.download('EA', period='1y', auto_adjust=True)
ticker 
```

```python
# Once we have the data we can create a function that appends the short MA 
# Typically, a MA is calculated using the closing price so we'll be adding the 'Close' column
shortMA = ticker['Close'].rolling(window=20).mean()
ticker['Short MA'] = shortMA
ticker
```

As we can see the first couple of values should be 'NaN' because we don't have the data of the previous 20 days, but now our dataframe has been automatically updated to have the short MA for each day. Now that we have this logic figured out, we can now construct our full functions to calculate and fill out new columns for short and long MA, for each stock in our portfolio. 

Our functions ended up looking like this: 

```python
def calculate_short_MA(ticker_data): 
    shortMA = ticker_data['Close'].rolling(window=25).mean()
    ticker_data['Short MA'] = shortMA
    
def calculate_long_MA(ticker_data): 
    longMA = ticker_data['Close'].rolling(window=100).mean()
    ticker_data['Long MA'] = longMA

# We can test out these with a stock from our portfolio 

ticker = yf.download('EA', start='2020-01-01', end='2025-01-01', auto_adjust=True)
calculate_short_MA(ticker)
calculate_long_MA(ticker)
ticker 
```

As we can see, our functions ended up working and now we have to create the logic in order to store where each moving average crosses or intersects. My first thought was to store a list of dates in which the averages would cross, but in order to create buy/sell signals we need to know the information around that date so instead I'm going to go ahead and decide to create booleans in order to act like a switch. Once that switch is triggered, then we know there was a crossover and it'll be easier to create the buy/sell logic. 

After playing around and trying to figure this out, I was trying to create a function in which it checks whether a boolean changes, and if it does for it to check the last 30 days to see if this was a whipsaw or a trend instead of a pattern, but I realize I can create much more simple logic if I create two more columns in each stock's data frame that check a simple boolean, and then can look at the data as a whole. Therefore, checking to see whether the switch had triggered multiple times in a short period would be easier, because it turns into a 'has this column been true for the last 30 days or has it flipped'. 

We start by creating these next two columns in our stock dataframe:

```python
buy_signal = ticker['Short MA'] > ticker['Long MA']
ticker['Buy Boolean'] = buy_signal
ticker['Sell Boolean'] = ~buy_signal
ticker
```

As we can see from the above graph, there are plenty of times where the sell signal should've caught a date but did not. Instead we'll have to fix our code before continuing, meaning that we are going to need to rework our buy and sell logic. I think that the dates caught in our code, were all crossovers rather than just buy signals, therefore I think there's something wrong with using the diff() method in our logic. 

After going over the data and analyzing each graph, I realized that our buy and sell logic wasn't grabbing the appropriate crossover dates. Instead of using the diff() method in pandas, I decided to make it more simpler and just add a new column in our dataframe called, 'Prev Bool', in which tracks what the previous boolean of the 'Switch' column was, in order to directly compare when in fact the switch signal converts from True to False or vice versa. This made our code shorter and it ended up looking like this: 

```python
def date_tracker(ticker): 
    buy_signal = ticker[(ticker['Prev Bool'] == True) & (ticker['Switch'] == False)]
    sell_signal = ticker[(ticker['Prev Bool'] == False) & (ticker['Switch'] == True)]
    print(buy_signal.index)
    print(sell_signal.index)

ticker = yf.download('EA', period='1y', auto_adjust=True)
calculate_short_MA(ticker)
calculate_long_MA(ticker)

def new_crossover(ticker): 
    switch_signal = ticker['Short MA'] > ticker['Long MA']
    ticker['Switch'] = switch_signal
    ticker['Prev Bool'] = ticker['Switch'].shift(1)

new_crossover(ticker)
date_tracker(ticker)
```
